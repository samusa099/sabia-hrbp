# Sabia Group HRBP Smartwatch Recovery 2026
Synthetic HR, production and financial analytics case.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

DATA = Path('/kaggle/input/sabia-group-hrbp-smartwatch-recovery-2026')
```

```python
score = pd.read_csv(DATA/'quarterly_scorecard.csv')
finance = pd.read_csv(DATA/'financial_impact_monthly.csv')
production = pd.read_csv(DATA/'production_monthly.csv')
score.head()
```

## Quarterly recovery scorecard

```python
display(score[['Metric','Unit','Q1','Q2','Q3','Q4','Annual_Target']])
```

## Monthly operating profit

```python
finance.plot(x='Month', y='Operating_Profit_BDT_Million', marker='o', figsize=(10,5))
plt.axhline(0, linewidth=1)
plt.xticks(rotation=45)
plt.tight_layout()
```

## Production quality by quarter

```python
production.groupby('Quarter').agg({'Actual_Units':'sum','First_Pass_Yield':'mean','Defect_Rate':'mean','Units_per_Labour_Hour':'mean'}).round(3)
```
